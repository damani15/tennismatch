### Define the functions

```python
#*****************GAME SCORE***************************
def game_score(currentServer):
    playerPointScore = [0,0] 
        
    #when the game is still going on (the game is over when a player reaches 4 or more points and wins by 2)
    while not(max(playerPointScore) >= 4 and abs(playerPointScore[0] - playerPointScore[1]) >= 2): 
        if random.random() < playerFirstServePercentage[currentServer]: #the server makes their first serve
            if random.random() < playerFirstServeWinPercentage[currentServer]: #if the server wins the point off the first serve
                playerPointScore[currentServer] += 1 
            else: #if the returner wins the point off the first serve
                playerPointScore[(currentServer + 1) % 2] += 1 # add a point to the other player (the returner)
        else: #the server misses their first serve
            if random.random() < playerSecondServeWinPercentage[currentServer]: #if the server wins the point off their second serve [****ISSUE 1: DID NOT ACCOUNT FOR DOUBLE FAULTS****]
                playerPointScore[currentServer] += 1
            else: #returner wins the point off their second serve
                playerPointScore[(currentServer + 1) % 2] += 1 # add a point to the other player (the returner)
                
    return playerPointScore
    
#***************TIEBREAK*************

def tiebreak_score(currentServer):
    playerTiebreakScore = [0,0]

    tiebreak_pointnumber = 1 # tracks the number of point in tiebreak
    tiebreak_currentServer = currentServer 
    #when the tiebreaker is still going on (the tiebreaker is over when a player reaches 7 or more points and wins by 2)
    while not(max(playerTiebreakScore) >= 7 and abs(playerTiebreakScore[0] - playerTiebreakScore[1]) >= 2): 
        if tiebreak_pointnumber % 4 == 1 or tiebreak_pointnumber % 4 == 0:
            tiebreak_currentServer = currentServer
        if tiebreak_pointnumber % 4 == 2 or tiebreak_pointnumber % 4 == 3:
            tiebreak_currentServer = (currentServer + 1) % 2
        if random.random() < playerFirstServePercentage[tiebreak_currentServer]: #the server makes their first serve
            if random.random() < playerFirstServeWinPercentage[tiebreak_currentServer]: #if the server wins the point off the first serve
                playerTiebreakScore[tiebreak_currentServer] += 1 #add point to the current server in tiebreak
            else: #if the returner wins the point off the first serve
                playerTiebreakScore[(tiebreak_currentServer + 1) % 2] += 1 # add a point to the returner
        else: #server misses first serve
            if random.random() < playerSecondServeWinPercentage[tiebreak_currentServer]: #server wins point off second serve [****ISSUE 1: DID NOT ACCOUNT FOR DOUBLE FAULTS****]
                playerTiebreakScore[tiebreak_currentServer] += 1
            else: #returner wins the point off their second serve
                playerPointScore[(tiebreak_currentServer + 1) % 2] += 1 # add a point to the returner 
        tiebreak_pointnumber +=1        
        
    #print('>>>> within tiebreak_score >>> exiting with score [%d,%d]' % (playerTiebreakScore[0],playerTiebreakScore[1]))
    return playerTiebreakScore
```

```python
import random

#(Nadal, Federer)

playerFirstServePercentage = [0.68, 0.62] #the percentage a player makes their first serve
playerFirstServeWinPercentage = [0.73, 0.77] #[player 0 first serve win percentage, player 1 first serve win percentage]
playerSecondServeWinPercentage = [0.54,0.59] #[player 0 second serve win percentage, player 1 second serve win percentage]

var = 100000
player1_wins = 0

for i in range(var):
    playerSetScore = [2,2]
    playerGameScore = [0,0]
    serveCounter = 0

    while max(playerSetScore) != 3: 
        playerPointScore = [0,0]
        playerTiebreakScore = [0,0]

        currentServer = (serveCounter % 2) # Give the remainder of serveCounter divided by 2, informs whether current server is player 1 or player 2

        #*********************SET SCORE************

        ##if the set is over (not including tiebreak), add a set to the player with the most games
        if (max(playerGameScore) == 6 or max(playerGameScore) == 7):
            if abs(playerGameScore[0] - playerGameScore[1]) >= 2: 
            if playerGameScore[0] - playerGameScore[1] > 0: #if player 0 has more games than player 1 in the set, add a set to player 0
                playerSetScore[0] += 1
                playerGameScore = [0,0]
            else: #do the opposite
                playerSetScore[1] += 1
                playerGameScore = [0,0]        

        ##if the game score is 6-6, run tiebreak function
        if playerGameScore[0] == 6 and playerGameScore[1] == 6: #if the score in the set is 6 all
            if playerSetScore[0] != 2 or playerSetScore[1] != 2:
                playerTiebreakScore = tiebreak_score(currentServer)
                if playerTiebreakScore[0] - playerTiebreakScore[1] > 0: #if player 0 has more games than player 1 in the set, add a set to player 0
                    playerSetScore[0] += 1
                    playerGameScore = [0,0]
                else: #do the opposite
                    playerSetScore[1] += 1
                    playerGameScore = [0,0]
            else: #keep playing until someone wins by 2
                playerPointScore = game_score(currentServer)
                if playerPointScore[0] - playerPointScore[1] > 0: 
                    playerGameScore[0] += 1
                else:
                    playerGameScore[1] += 1
        ##if the set is not over, play another game
        else:
            playerPointScore = game_score(currentServer)
            if playerPointScore[0] - playerPointScore[1] > 0: 
                playerGameScore[0] += 1
            else:
                playerGameScore[1] += 1

        serveCounter += 1

        print('Points in Game: %s' % str(playerPointScore))
        print('Games in Set:   %s' % str(playerGameScore))
        print('Score of Sets:  %s' % str(playerSetScore))
        print()

    #******************MATCH******************

 ##if the match is over, print winner (player 0 or player 1)
    if playerSetScore[1] - playerSetScore[0] > 0: #if player 1 has more set than player 0 in the set, add a set to player 1
        #print ("Winner: Player 1")
        player1_wins += 1

print (player1_wins/var*100)
    
```
